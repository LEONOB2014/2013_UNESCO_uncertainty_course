```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#READ in a csv file, only the theta columns, dates are parsed from file, the days are in front of the months
meas = pd.read_csv('showcase_seminar.csv', index_col=0, header=0, parse_dates=True, dayfirst=True)
#give the columns some logical names
meas.columns=['Node 10','Node 20','Node 30','Node 40','Node 50']
#check first rows
meas.head()
```

```python
#check some statistics about the 
meas.describe()
```

```python
#plot the dataframe (matplotlib and pandas)
meas.plot(subplots=True, figsize=(10, 8), yticks=[0.0,20.,40.])
```

```python
pd.rolling_mean(meas, 200).plot()
```

```python
meas.resample('D', how='mean').plot()
```
